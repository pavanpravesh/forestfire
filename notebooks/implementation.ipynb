In pandas, .loc[] and .iloc[] are both used for accessing data in a DataFrame, but they differ in how they select rows and columns.

.loc[]:
Label-based indexing.

You use row and column labels to select data.

Includes both the start and stop labels (the end is inclusive).

Example:

python
Copy code
df.loc[2:4, 'A':'C']  # Selects rows with index labels 2, 3, 4 and columns 'A', 'B', 'C'
.iloc[]:
Integer-based (position-based) indexing.

You use row and column integer positions to select data.

Does not include the stop index (the end is exclusive).

Example:

python
Copy code
df.iloc[2:4, 0:3]  # Selects the 2nd and 3rd rows (positions) and the first 3 columns
Summary:
.loc[]: Select by labels and includes the last label.
.iloc[]: Select by positions and excludes the last index.
